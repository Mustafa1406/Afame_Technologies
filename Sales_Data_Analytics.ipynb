```python
ax = sns.countplot(x = 'Category', data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

From above analysis we can see that count-wise sales of office supplies are more than technology and furniture


```python
df.groupby(['Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
```

```python
sales_by_cat = df.groupby(['Category'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
sns.barplot(x='Category', y='Sales', data = sales_by_cat)
```

From above graph we can contemplate that with respect to sales technology has more sales rather than furniture and office supplies.


```python
# Revenue over time
df['revenue'] = df['Sales'] * df['Quantity']
df.set_index('Order Date', inplace=True)
revenue_per_date = df.resample('D')['revenue'].sum()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=revenue_per_date)
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

The above graph shows the sales within the duration of every 6 months initiating from 2011 to 2015.
Additionally, in 2014 highest number of sales was done.


```python
df.columns
```

```python
# Total revenue by sub category
total_revenue_by_sub_category = df.groupby('Sub-Category')['revenue'].sum().reset_index().sort_values(by='revenue', ascending=False)
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(data=total_revenue_by_sub_category, x='Sub-Category', y='revenue', estimator=sum)
plt.title('Total Revenue by Sub_Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 10), textcoords = 'offset points')

plt.show()
```

From above chart we can see that maximum sales is done by Phones,then chairs and copiers and so on.


```python
df.columns
```

```python
# Best selling products
best_selling_products = df.groupby('Product Name')['revenue'].sum().reset_index().sort_values(by='revenue', ascending=False)
print("Best-Selling Products by Revenue:")
print(best_selling_products)
```

**Computing total sales**


```python
total_sales = df['revenue'].sum()
print(f"Total Sales Revenue: ${total_sales:,.2f}")
```

**Analyzing Sales trends over time**

```python
monthly_sales = df['revenue'].resample('M').sum()

plt.figure(figsize=(12, 7))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(False)
plt.show()
```

**Determine best selling products**

```python
total_revenue_by_product = df.groupby('Product Name')['revenue'].sum().reset_index()

best_selling_products = total_revenue_by_product.sort_values(by='revenue', ascending=False)

top_10_products = best_selling_products.head(10)
print(top_10_products)
```

**Visualize the Data**

```python
plt.figure(figsize=(10, 5))
bar_plot = sns.barplot(data=top_10_products, x='Product Name', y='revenue', estimator=sum)
plt.title('Top 10 Best-Selling Products by Revenue')
plt.xlabel('Products')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

# Adding data labels
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center',
                      xytext=(0, 10), textcoords='offset points')

plt.show()
```
